#### Importing data directly from built in pycaret dataset

```
import pandas as pd
```

```
from pycaret.datasets import get_data

index = get_data('index')
```

```
data = get_data('automobile')
```

#### Initializing Set-up

```
from pycaret.regression import *
reg = setup(data, target = 'price', session_id=123, log_experiment=True, experiment_name='automobile')
 
```

### Checking and filling missing values

```
data.isnull().sum()
```

```
data['normalized-losses'] = data['normalized-losses'].fillna(data['normalized-losses'].mean())
data['num-of-doors'] = data['num-of-doors'].fillna(data['num-of-doors'].mode()[0])

data.isnull().sum().sum()
```

#### Comparing baseline models

```
best_model = compare_models(fold=5)
```

## Build the Model

```
gbr = create_model('gbr')
```

### Tuning the model

```
tuned_gbr = tune_model(gbr, n_iter=50, optimize = 'MAE')
```

```
print(tuned_gbr)
```

### Plot Model

```
plot_model(gbr)
```
